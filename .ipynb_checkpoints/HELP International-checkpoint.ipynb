```python
# Reducing the data
```

```python
df= data.copy()
```

```python
del df['country']
```

```python
#Scaling the data
sc=StandardScaler()
df_s=pd.DataFrame(sc.fit_transform(df),columns=df.columns)
```

```python
df_s
```

```python
#Dimension Reduction
pca=PCA(n_components=9)
pca_data=pca.fit_transform(df_s)
```

```python
#Eigen Values
pca.explained_variance_
```

```python
#Eigen Vectors
pca.components_
```

```python
#Explained Variance Ratio
np.cumsum(pca.explained_variance_ratio_)
```

```python
#Since we are targeting an explained variance of close to 95% , we will take the number of components as 5.
```

```python
pca=PCA(n_components=5)
pca_data=pca.fit_transform(df_s)
```

```python
df_pca=pd.DataFrame(pca_data,columns=['PC1','PC2','PC3','PC4','PC5'])
```

```python
df_pca
```

```python
#Finding the right number of clusters
WCSS=[]  #-----Distortion Scores
for i in range(2,11):
    model=KMeans(n_clusters=i)
    pred=model.fit_predict(df_pca)
    WCSS.append(silhouette_score(df_pca,pred))
```

```python
WCSS
```

```python
#Visualizing the distortion scores to get the number of clusters.
plt.plot(range(2,11),WCSS)
```

```python
#Using built-in functions to visualize.
kv=KElbowVisualizer(KMeans(),k=10)
kv.fit(df_s)
kv.show()
```

```python
#As is evident the optimum number of clusters are 4, we will still check using Silhoutte Score.
```

```python
#Silhoutte Score
sil_score=[]
for i in range(2,11):
    k=KMeans(n_clusters=i)
    pred=k.fit_predict(df_pca)
    sil_score.append(silhouette_score(df_pca,pred))
    
```

```python
sil_score  #Silhoutte Score for the four clusters is the highest.
```

```python
#Visualizing the Silhoutte Score.
t=1
plt.figure(figsize=(12,20))
for i in range(2,11):
    plt.subplot(6,2,t)
    k=KMeans(n_clusters=i)
    sil=SilhouetteVisualizer(k)
    sil_viz=sil.fit(df_pca)
    t+=1
    plt.title(f'k:{i},Silhoutte Score:{np.round(sil_viz.silhouette_score_,2)} ')
plt.tight_layout()
plt.show()
    
    
    
```

```python
#Clustering using K-Means
k=KMeans(n_clusters=4)
pred=k.fit_predict(df_pca)
```

```python
#Adding the clusters to the DataFrame
data['labels']=pred
```

```python
#Visualizing the clusters
sns.scatterplot(x=df_pca['PC1'],y=df_pca['PC2'],hue=data['labels'])
plt.show()
```

```python
#Datapoints in the first cluster.
data.loc[data['labels']==0] 
```

```python
#Since the cluster has only three datapoints, without any vsisble relationship we will drop the cluster,
#And make new clusters,with no.of clusters equal to 3.
```

```python
k=KMeans(n_clusters=3)
pred=k.fit_predict(df_pca)
```

```python
data['labels']=pred
```

```python
sns.scatterplot(x=df_pca['PC1'],y=df_pca['PC2'],hue=data['labels'])
plt.show()
```

```python
#Using groupby to understand the nature of Clusters.
data.groupby(by='labels').mean()
```

#### Inferences:
* Countries under the first cluster(Label 0) are developed countries,this is evident by a high GDP per capita(gdpp) and strong health indicators like low child mortality rate(child_mort),high life expectancy(life_expec), low total fertility rate(total_fer).
**These countries are not to be considered for Aid**.

* Countries under the second cluster(Label 1) are under-developed countries,this is evident by a low GDP per capita(gdpp) and weak health indicators like high child mortality rate(child_mort),low life expectancy(life_expec), high total fertility rate(total_fer).
**These countries should be the first priority for Aid**.

* Countries under the third cluster(Label 2) are mid-developed countries,this is evident by a comparatively high GDP per capita(gdpp) and comparatively good health indicators like low to hig child mortality rate(child_mort),high life expectancy(life_expec), low total fertility rate(total_fer).
**These countries may be considered for Aid,depending upon the corpus utilization for the under-developed category**.



```python
#Assigning classes according to the inferences.
data['Class'] = data['labels'].map({0:'Rich',1:'Poor',2:'Average'})
```
